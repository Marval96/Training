```python
#Llamado de librerías
import numpy as np
import matplotlib.pyplot as plt

# Generar dos matrices de números aleatorios entre 0 y 1 (coordenadas)
x = np.random.rand(100)
y = np.random.rand(100)

# Calcular la distancia de cada punto al origen (0, 0)
distancia = np.sqrt(x**2 + y**2)

# Contar el número de puntos que están a una distancia menor o igual a 1
puntos = np.sum(distancia <= 1)

# Calcular la probabilidad como el cociente entre el número de puntos dentro del círculo 
# y el número total de puntos
prob = puntos / len(x)

# Imprimir la probabilidad y el número pi
print("Probabilidad:", prob)
print("Número π:", prob * 4)

# Visualizar los puntos en un gráfico
plt.scatter(x, y, c=distancia <= 1, cmap='binary')
plt.show()
```

```python
# Llamado de librerias 
import numpy as np
import matplotlib.pyplot as plt

# Generar n pares de números aleatorios (x,y)
def generate_points(n):
    return np.random.rand(n, 2)

# Calcular la distancia entre dos puntos
def distance(p1, p2):
    return np.sqrt((p1[0]-p2[0])**2 + (p1[1]-p2[1])**2)

# Calcular la probabilidad de que un punto se encuentre a una distancia menor o igual a 1 del punto (0,0)
def calculate_probability(points):
    count = np.sum([distance(point, [0,0]) <= 1 for point in points])
    return count / len(points)

# Calcular 4P para diferentes valores de n
n_values = np.arange(1, 10000, 100)
pi_estimates = []
for n in n_values:
    points = generate_points(n)
    probability = calculate_probability(points)
    pi_estimates.append(4 * probability)

# Graficar el valor 4P frente al n
plt.plot(n_values, pi_estimates)
plt.axhline(y=np.pi, color='r', linestyle='--')
plt.xlabel('n')
plt.ylabel('4P')
plt.title('Estimación de Pi mediante números aleatorios')
plt.show()
```
